```python
# List the names of all episodes in which both Jerry Smith and Summer Smith appeared.
jerry = char[char['name'] == 'Jerry Smith']['episode'][4]
summer = char[char['name'] == 'Summer Smith']['episode'][2]

intr = set(summer).intersection(set(jerry))
for i in intr:
    print(requests.get(i).json()['name'])
```
